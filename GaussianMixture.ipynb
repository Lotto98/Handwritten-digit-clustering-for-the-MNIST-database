Mean Plot for Gaussian Mixture

```python
import matplotlib.pyplot as plt
from math import ceil
import pickle

def plot_means(n_pc:int, means)->None:
    """
    Function that given a PCA dimension and the mean of each Gaussian, plots the PCA reconstructed mean for each Cluster

    Args:
        n_pc (int): pca dimension.
        means (): mean of each Gaussian.
    """
    
    with open("pca/pca_"+str(n_pc)+".pkl", 'rb') as inp:
        pca=pickle.load(inp)
        
    data=pca.inverse_transform(means)
    
    fig,axs=plt.subplots(ceil(len(means)/4),4)
    
    fig.suptitle(f"Means plot for PCA dimension {n_pc}, with {len(means)} clusters:")
    
    axs = [item for sublist in axs for item in sublist]
    
    for i in range(len(axs)):
        axs[i].axis('off')
        
    for i,mean in enumerate(data):
        axs[i].imshow(mean.reshape(28, 28))

for x in [best_dim, worst_dim]:
    plot_means(x,fitted_estimators[x].means_)
```
